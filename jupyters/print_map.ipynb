```python
import folium
from folium.plugins import HeatMap
from os import system
import src.create_geoindex as geo
import pandas as pd

map_city=folium.Map(location=[37.3382,-121.8863], zoom_start=12)

collections = ["airport_sanjose_ca","elementary_school_sanjose_ca","kindergarten_sanjose_ca","night_clubs_sanjose_ca","starbucks_sanjose","train_stations_sanjose_ca","vegan_restaurants_sanjose_ca"]    
                            
```

```python
db, coll = geo.connectCollection("companies","airport_sanjose_ca")
places = list(coll.find({"geometry.location.lat":{"$ne":None},"geometry.location.lng":{"$ne":None}}))

for place in places:
    folium.Marker(place['location']['coordinates'][::-1],
                radius=2,
                icon=folium.Icon(icon='plane',color='blue'), 
    ).add_to(map_city) 
```
